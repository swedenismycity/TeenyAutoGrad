```python
import random

class Tree:
    def __init__(self):
        self.value = str(random.randint(0,1000))
        self._child = []
    def add_parr(self, end, level=0):
        if end < level:
            return
        self._child = [Tree(), Tree()]
        for c in self._child:
            c.add_parr(end, level=level+1)
```

```python
t = Tree()
t.add_parr(2)
```

```python
from graphviz import Digraph
import os
os.environ["PATH"] += os.pathsep + "C:\\Program Files\\Graphviz\\bin"


dot = Digraph() 

dot.attr(rankdir='LR')
#dot.attr(rankdir='RL'); dot.attr('edge', dir='back')

dot.attr('node', shape='box')

def rec(n):
    dot.node(n.value, n.value)
    for c in n._child:
        dot.node(c.value, c.value)
        dot.edge(n.value, c.value)
        rec(c)
rec(t)
```
